```python
import requests
from datetime import datetime, timedelta


lat = 49.8728  # Latitude of Darmstadt, Germany
lon = 8.6512   # Longitude of Darmstadt, Germany


start_date = "2023-04-11"
end_date = "2023-09-13"


api_key = "bd03a812ee86e15ea431c5cd7c03b6a1"


start_datetime = datetime.strptime(start_date, "%Y-%m-%d")
end_datetime = datetime.strptime(end_date, "%Y-%m-%d")


endpoint_url = "https://api.openweathermap.org/data/3.0/onecall/timemachine"


current_datetime = start_datetime


weather_data = []

while current_datetime <= end_datetime:
    
    current_unix_timestamp = int(current_datetime.timestamp())

    
    url = f"{endpoint_url}?lat={lat}&lon={lon}&dt={current_unix_timestamp}&appid={api_key}"

    
    response = requests.get(url)

    
    if response.status_code == 200:
        data = response.json()
        
        
        temperature_celsius = data['data'][0]['temp'] - 273.15  # Convert Kelvin to Celsius
        weather_main = data['data'][0]['weather'][0]['main']

        
        weather_data.append({
            'date': current_datetime.strftime('%Y-%m-%d'),
            'temperature_celsius': round(temperature_celsius, 2),
            'weather': weather_main
        })

        
        print(f"Date: {current_datetime.strftime('%Y-%m-%d')}, Temperature (Celsius): {temperature_celsius:.2f}°C, Weather: {weather_main}")
    else:
        print(f"Error for {current_datetime.strftime('%Y-%m-%d')}: {response.status_code} - {response.text}")

    
    current_datetime += timedelta(days=1)
```
